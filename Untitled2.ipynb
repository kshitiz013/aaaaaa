```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Constants
hbar = 1.0545718e-34  # Reduced Planck's constant (JÂ·s)
m = 9.10938356e-31  # Mass of an electron (kg)
e = 1.602176634e-19  # Elementary charge (C)
```

```python
# Function to calculate tunneling probability
def tunneling_probability(V0, a, E):
    if E < V0:
        kappa = np.sqrt(2 * m * (V0 - E)) / hbar
        T = np.exp(-2 * kappa * a)
    else:
        T = 1  # Full transmission when E >= V0
    return T
```

```python
# Parameter ranges
V0_values = np.linspace(1, 5, 5) * e  # Barrier heights (J)
a_values = np.linspace(1e-10, 5e-10, 5)  # Barrier widths (m)
E_values = np.linspace(0.1, 5, 50) * e  # Particle energies (J)

# Generate dataset
data = []
for V0 in V0_values:
    for a in a_values:
        for E in E_values:
            T = tunneling_probability(V0, a, E)
            data.append({"Barrier_Height_J": V0, "Barrier_Width_m": a, "Energy_J": E, "Tunneling_Probability": T})

# Convert to DataFrame
df = pd.DataFrame(data)
```

```python
# Save the dataset
df.to_csv("quantum_tunneling_data.csv", index=False)

# Visualization
plt.figure(figsize=(10, 6))
for a in a_values:
    T_values = [tunneling_probability(V0_values[0], a, E) for E in E_values]
    plt.plot(E_values / e, T_values, label=f"Width = {a:.1e} m")

plt.title("Tunneling Probability vs Particle Energy")
plt.xlabel("Particle Energy (eV)")
plt.ylabel("Tunneling Probability")
plt.legend()
plt.grid(True)
plt.show()
```
